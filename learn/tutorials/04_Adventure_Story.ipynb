## The Magic Forest

Run this code to start your adventure:

```python
print("="*50)
print("  THE MAGIC FOREST")
print("  A Choose Your Own Adventure")
print("="*50)
print()
print("You wake up in a magical forest.")
print("The trees glow with soft colors.")
print("You see two paths ahead:")
print()
print("  1. A path with purple glowing mushrooms")
print("  2. A path with a sparkling stream")
print()
```

```python
choice = input("Which path do you choose? (1 or 2): ")

if choice == "1":
    print()
    print("You follow the mushroom path...")
    print("A tiny fairy appears!")
    print("'Hello traveler!' she says.")
    print("'I can grant you a wish!'")
elif choice == "2":
    print()
    print("You walk beside the sparkling stream...")
    print("A friendly dragon is drinking water!")
    print("It looks at you and smiles.")
    print("'Would you like a ride?' it asks.")
else:
    print("You got lost in the forest!")
    print("Try again with 1 or 2.")
```

## Making It Bigger!

Let's add more choices to our story:

```python
def magic_forest_adventure():
    print("="*50)
    print("  THE MAGIC FOREST - FULL ADVENTURE")
    print("="*50)
    print()
    
    # Get the player's name
    name = input("What is your name, brave adventurer? ")
    print()
    print(f"Welcome, {name}!")
    print()
    print("You wake up in a magical forest.")
    print("The trees glow with soft colors.")
    print()
    print("You see two paths:")
    print("  1. Purple mushroom path")
    print("  2. Sparkling stream path")
    print()
    
    path = input("Choose your path (1 or 2): ")
    
    if path == "1":
        mushroom_path(name)
    else:
        stream_path(name)

def mushroom_path(name):
    print()
    print("You follow the mushroom path...")
    print("A tiny fairy appears in a flash of light!")
    print(f"'Hello {name}! I'm Sparkle,' she says.")
    print("'I can help you! Do you want:'")
    print("  1. A magic wand")
    print("  2. A bag of gold")
    print()
    
    wish = input("Choose your gift (1 or 2): ")
    
    if wish == "1":
        print()
        print("Sparkle gives you a beautiful wand!")
        print("It glows with rainbow colors!")
        print(f"'{name}, you can now do ONE magic spell!'")
        print()
        print("THE END - You became a wizard!")
    else:
        print()
        print("Sparkle gives you a heavy bag of gold coins!")
        print(f"'{name}, use it wisely,' she says.")
        print()
        print("THE END - You're rich!")

def stream_path(name):
    print()
    print("You walk beside the sparkling stream...")
    print("A friendly blue dragon is there!")
    print(f"'Oh! A human! Hi {name}!'")
    print("'I'm Bluebell. Want to:'")
    print("  1. Fly on my back")
    print("  2. Swim in the magic water")
    print()
    
    choice = input("What do you do? (1 or 2): ")
    
    if choice == "1":
        print()
        print("You climb on Bluebell's back!")
        print("WHOOOOSH! You fly up into the clouds!")
        print("You can see the whole magic forest!")
        print()
        print("THE END - You made a dragon friend!")
    else:
        print()
        print("You jump into the sparkling water...")
        print("You can breathe underwater!")
        print("You see a beautiful underwater castle!")
        print()
        print("THE END - You found a secret kingdom!")

# START THE GAME!
magic_forest_adventure()
```
