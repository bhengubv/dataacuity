# Make a Guessing Game! 

Let's build a game where the computer picks a secret number and you try to guess it!

## The Computer's Random Number

First, let's learn how to make the computer pick a random number:

```python
import random

# Pick a random number from 1 to 10
secret_number = random.randint(1, 10)

print("I'm thinking of a number between 1 and 10...")
print("(Psst, for testing, the secret is:", secret_number, ")")
```

## Getting Your Guess

Now let's ask the player to guess:

```python
# Ask for a guess
guess = int(input("What's your guess? "))
print("You guessed:", guess)
```

## Checking the Guess

Now we need to tell the player if they're right, too high, or too low:

```python
if guess == secret_number:
    print("YOU WIN! The number was", secret_number)
elif guess > secret_number:
    print("Too HIGH! Try a smaller number.")
else:
    print("Too LOW! Try a bigger number.")
```

## The Complete Game!

Let's put it all together with multiple tries:

```python
import random

print("="*40)
print("  WELCOME TO THE GUESSING GAME!")
print("="*40)
print()

# Pick a secret number
secret_number = random.randint(1, 10)
tries = 0
won = False

print("I'm thinking of a number between 1 and 10.")
print("You have 3 tries to guess it!")
print()

# Give 3 tries
for attempt in range(3):
    guess = int(input(f"Try {attempt + 1}: What's your guess? "))
    tries += 1
    
    if guess == secret_number:
        print()
        print("YOU WIN! You got it in", tries, "tries!")
        won = True
        break
    elif guess > secret_number:
        print("Too high! Try lower.")
    else:
        print("Too low! Try higher.")

if not won:
    print()
    print("Game Over! The number was", secret_number)
    print("Better luck next time!")
```
