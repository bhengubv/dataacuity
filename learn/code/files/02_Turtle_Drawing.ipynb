# Drawing with Turtle! 

Meet your new friend: **Turtle**! It's a little arrow that can draw pictures when you tell it where to go.

## First, Let's Meet Turtle

Run this code to bring Turtle to the screen:

```python
import turtle

# Create a screen and a turtle
screen = turtle.Screen()
t = turtle.Turtle()

# Say hello!
print("Turtle is ready to draw!")
```

## Making Turtle Move

Turtle can:
- `forward(steps)` - Go forward
- `right(degrees)` - Turn right
- `left(degrees)` - Turn left

### Let's Draw a Line!

```python
t.forward(100)  # Go forward 100 steps
```

### Draw a Square!

A square has 4 sides and 4 turns. Each turn is 90 degrees (a right angle).

```python
# Clear the screen first
t.clear()
t.home()  # Go back to center

# Draw a square
t.forward(100)
t.right(90)
t.forward(100)
t.right(90)
t.forward(100)
t.right(90)
t.forward(100)
```

## Using Loops - The Easy Way!

Instead of writing `forward` and `right` 4 times, we can use a **loop**:

```python
t.clear()
t.home()

# Do this 4 times:
for i in range(4):
    t.forward(100)
    t.right(90)
```

## Adding Colors! 

Let's make things colorful!

```python
t.clear()
t.home()

# Set the pen color
t.pencolor("red")
t.forward(100)

t.pencolor("green")
t.right(90)
t.forward(100)

t.pencolor("blue")
t.right(90)
t.forward(100)

t.pencolor("orange")
t.right(90)
t.forward(100)
```

## Super Challenge: Make a Rainbow Spiral!

```python
t.clear()
t.home()
t.speed(0)  # Super fast!

colors = ["red", "orange", "yellow", "green", "blue", "purple"]

for i in range(100):
    t.pencolor(colors[i % 6])  # Pick a color
    t.forward(i * 2)  # Go forward (longer each time)
    t.right(60)  # Turn
```

## What You Learned

- Turtle is a fun way to draw with code!
- `forward(steps)` moves turtle forward
- `right(degrees)` turns turtle right
- `for i in range(n):` repeats code n times
- `pencolor("color")` changes the color
